```python
import pandas as pd
```

```python

train_categorical_percent = pd.read_csv('D:/Project/Sample data/Sam_8/train_categorical.csv')
```

```python
train_numeric_percent = pd.read_csv('D:/Project/Sample data/Sam_8/train_numeric_1000k.csv')
```

```python
ifw_percent = pd.read_csv('D:/Project/Sample data/Sam_8/label_10molding_machine_5000dp.csv')
```

```python
ifw_percent
```

```python
# Train categorical percent of missing Value 

percent_missing = ifw_percent.isnull().sum() * 100 / len(ifw_percent)
missing_value_df = pd.DataFrame({'column_name': ifw_percent.columns,
                                 'percent_missing': percent_missing})

col_num=len(ifw_percent.columns)
Total = missing_value_df['percent_missing'].sum()
print("IFW_percent:=""%.0f%%" % (Total/col_num))
```
