```python
import numpy as np
from scipy.io import loadmat  # this is the SciPy module that loads mat-files
import matplotlib.pyplot as plt
from datetime import datetime, date, time
import pandas as pd

mat = loadmat('C:/Users/Rezaul Abedin/Desktop/T1.mat')
df = pd.DataFrame(mat['T1'])


import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
df_sample = df.sample(n=round(43872950*.01), random_state=1)

Data = {'x': df_sample[0].index.values,
        'y': df_sample[0].values
       }
  
df_new = pd.DataFrame(Data,columns=['x','y'])
  
kmeans = KMeans(n_clusters=3).fit(df_new)
centroids = kmeans.cluster_centers_
print(centroids)

plt.figure(figsize=(30,30))

plt.scatter(df_new['x'], df_new['y'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)

plt.savefig('D:/Project_data_all_L3S/Process_data/KM0.png')
plt.show()
```
