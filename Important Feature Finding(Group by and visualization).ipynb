```python
#Feature group by and visualization.
import pandas as pd
import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Important Feature Numeric
path = r'D:/Important Features Num/input output csv_10new300' # useing my csv path
all_files = glob.glob(path + "/*.csv")

list_csv = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=False, usecols=["f_name", "score"])
    list_csv.append(df)
    #print (list_csv)

df = pd.concat(list_csv, axis=0, ignore_index=False)
```

```python
df
```

```python
data=df.groupby(['f_name'], as_index=False).mean()
data=data.sort_values(by=['score'], ascending=False)
```

```python
data ["score"] = data ["score"]*100
```

```python
data
```

```python
#feature_name =data.fe
feature_name =data["f_name"]
score =data["score"]

inter_2 = feature_name.head(30).tolist()
feature_name = feature_name.head(30).tolist()
score = score.head(30).tolist()
```

```python
inter_2[0:5]
```

```python
feature_name[0:5]
```

```python
y_pos = np.arange(len(feature_name))
fig, ax = plt.subplots(figsize=(10, 5)) 
# Create bars and choose color
plt.bar(y_pos, score, color = (0.5,0.1,0.5,0.6))
 
# Add title and axis names
plt.title('Important Feature')
plt.xlabel('Feature Name')
plt.ylabel('Score')
 
# Limits for the Y axis
plt.ylim(0,12)
 
# Create names
plt.xticks(y_pos, feature_name,rotation=90)
 
# Show graphic
plt.show()
```
