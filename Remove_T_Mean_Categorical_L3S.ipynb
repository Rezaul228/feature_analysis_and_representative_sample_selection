```python
import csv
import gc
def get_data(directory, file_list, chunksize, data_type_list = None, use_col_list = None):
    # type: (str, list, list, list) -> pd.DataFrame
    """
    :param directory: input directory location
    :param file_list: list of files to read
    :param data_type_list: list of data types for each file (ex. int, str, etc)
    :param use_col_list: list of columns to read from
    :return: dataframe of all files
    """

    data = None
    for i, file_name in enumerate(file_list):
        #print (file_name)
        subset = None
        for j, chunk in enumerate(
                pd.read_csv(directory + file_name, chunksize=chunksize, low_memory=False, skiprows= np.arange(200000, 1186747) )):
            #print (j)
            if subset is None:
                subset = chunk.copy()
            else:
                subset = pd.concat([subset, chunk])
            del chunk
            gc.collect()
        if data is None:
            data = subset.copy()
        else:
            data = pd.merge(data, subset.copy(), on="Id")
        del subset
        gc.collect()

    return data
```

```python
import pandas as pd 
import pickle

in_path = "D:/Project_data_all_L3S/train_categorical_T.csv" #Path where the large file is
out_path = "D:/Project_data_all_L3S/" #Path to save the pickle files to
chunk_size = 20000 #size of chunks relies on your available memory
separator = ","

reader = pd.read_csv(in_path,sep=separator,chunksize=chunk_size, 
                    low_memory=False)    


def dropT(chunk):
    chunk=chunk.replace('T','',regex=True)
    return chunk

for i, chunk in enumerate(reader):
    out_file = out_path + "/loaddata_{}.pkl".format(i+1)
    with open(out_file, "wb") as f:
        pickle.dump(dropT,f,pickle.HIGHEST_PROTOCOL)
```
