```python
# if feature name match with another score will add in score1 column, or the respective row will add as new row in bottom.
import math
import numpy as np
import pandas as pd
import glob
import matplotlib.pyplot as plt
from collections import Counter
import re
all_data_ipf1 = pd.read_csv("C:/Users/Rezaul Abedin/Desktop/chunk/output_chunk_fifty/New folder/New folder/data_50.csv", nrows=10,index_col=0 )
all_data_ipf2 = pd.read_csv("C:/Users/Rezaul Abedin/Desktop/chunk/output_chunk_fifty/New folder/New folder/data_1.csv", nrows=10,index_col=0 )
```

```python
two_score_data = all_data_ipf1.copy()
two_score_data['score1'] = 0

for sort_name in all_data_ipf2.f_name.values:
    #print ("one row complete")
    for sort_according in all_data_ipf1.f_name.values:
        #print (sort_name)
        flag= 0
       
        if sort_according == sort_name:
            flag = 1
            sort_name_row = all_data_ipf2.loc[all_data_ipf2['f_name'] == sort_name]
            name = sort_name_row.iloc[0]['f_name']
            value = sort_name_row.iloc[0]['score']
            #print(name)
            #print(value)
            two_score_data['score1'] = np.where(two_score_data['f_name'] ==str(name), value, two_score_data['score1'])
            
    if(flag == 0):
        sort_name_row_n = all_data_ipf2.loc[all_data_ipf2['f_name'] == sort_name]
        name_n = sort_name_row_n.iloc[0]['f_name']
        value_n = sort_name_row_n.iloc[0]['score']
        #print(sort_name_row_n)
        two_score_data = two_score_data.append({'f_name': name_n, 'score': 0, 'score1': value_n}, ignore_index=True)
        
print ( two_score_data)
```
